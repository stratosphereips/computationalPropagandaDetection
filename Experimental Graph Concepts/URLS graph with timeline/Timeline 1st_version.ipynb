```python
for idx in range(1, 3):

    # load database 
    url = all_urls[idx]
    db = DB(f"C:\\propaganda\\computationalPropagandaDetection\\DB\\v1\\propaganda.{idx}.db")
    edges, nodes = db.get_tree(url)


    node_names = [n for n in list(nodes.keys()) if n != url]
    links_without_level = [(from_id, to_id) for (_, from_id, to_id, _) in edges if from_id != url]
    possible_colors = ["red", "green", "c", "m", "y"]
    dates = [str(nodes[url][0][:10]) if nodes[url][0] is not None else None for url in node_names]
    unique_dates = sorted(list({str(d) for d in dates}))
    if 'None' not in unique_dates:
        unique_dates.append('None')

    pos = [(datetime.datetime.strptime(date, '%Y-%m-%d').date() - datetime.date.today()).days for date in unique_dates[:-1]]


    edges_to_date = [(n, nodes[n][0][:10]) for n in node_names if nodes[n][0] is not None]



    date_x = {}
    for i, n in enumerate(unique_dates[:-1]):
        date_x[n] = int((pos[i]- sum(pos)/len(pos)))
    edge_dict = dict()

    for level ,source, target, _ in edges:
        if source not in edge_dict:
            edge_dict[source] = []
        edge_dict[source].append(target)

    colors=["red", "blue", "green"]



    g = Network('700px', '700',notebook=False, directed=False)

    for n in node_names:
        g.add_node(n, label = filter_name(n), color=colors[nodes[n][1]])


    for e in links_without_level:
        g.add_edge(*e, physics=True, hidden=True)

        # for some reason position of date is broken
    for i, n in enumerate(unique_dates[:-1]):
        g.add_node(n, label=' ', title=n, color='gray', x=date_x[n], y=0, physics=True, shape='diamond', size=40, fixed={'x':True,'y':True})

    # different physics simulations
    # g.force_atlas_2based()
    # g.repulsion()


    for e in edges_to_date:
        g.add_edge(*e, color='gray', physics=False)

    g.set_edge_smooth('smooth')
    g.show(f'Notebook graphs\\\Timeline graphs\\1st_version_{idx}.html')
```
