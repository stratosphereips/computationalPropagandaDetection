```python
# Changable parameters

TIME_WINDOW = 2 # Cluster older days than TIME_WINDOW ago into single node
fix_level = True  # Fix level 1 nodes in the y-axis
date_edge_physics = True  # If nodes should be pulled to the date 
k=8
```

```python
for idx in range(1, 41):
    url = all_urls[idx]
    db = DB(f"DB/v1/propaganda.{idx}.db")
    edges, nodes = db.get_tree(url)
    query_date = datetime.datetime.strptime(nodes[url][0][:10], '%Y-%m-%d').date() if nodes[url][0] is not None else None

    dates = [str(nodes[url][0][:10]) if nodes[url][0] is not None else None for url in nodes]
    dates.append(None)
    unique_dates = sorted(list({str(d) for d in dates}))

    date_dict = dict()
    last_date = datetime.date.today()
    first_date = datetime.date.today() - datetime.timedelta(days=TIME_WINDOW*365)
    pos = {d:(datetime.datetime.strptime(d, '%Y-%m-%d').date() - first_date).days for d in unique_dates[:-1]}
    if query_date is not None and first_date is not None:
        query_date_position = k*(query_date-first_date).days
    else:
        query_date_position = None


    centers = {s for lvl ,s,_, _ in edges if lvl == 2}
    clusters = {}
    for c in centers:

        clusters[c] = {'links':[], 'position':[]}
        clusters[c]['links'] = [(target, source_type) for _,source,target,source_type in edges if source == c]


    colors=["red", "blue", "green"]
    # initialize the graph
    g = Network('700px', '1500',notebook=False, directed=False)

    # add the timeline, need to skip the query date and add later (because of the coloring)
    for i, n in enumerate(unique_dates[:-1]):
        if query_date is None or n != query_date and n is not None:
            date = datetime.datetime.strptime(n, '%Y-%m-%d').date()
            if k*(date-first_date).days == query_date_position:
                g.add_node(str(query_date), label=' ', title='original date' + str(query_date), color="red", x=query_date_position, y=0, physics=True, shape='diamond', size=25, fixed={'x':True,'y':True})

                continue

            if first_date < date < last_date:
                g.add_node(n, label=' ', title=n, color="#cccccc", x=k*(date-first_date).days, y=0, physics=False, shape='diamond', size=20, fixed={'x':True,'y':True})
            elif date >= last_date:
                g.add_node(n, label=' ', title=n, color="#555555", x=k*(date-first_date).days, y=0, physics=False, shape='diamond', size=15, fixed={'x':True,'y':True})

    g.add_node(str(datetime.date.today()), label=' ', title='Query date ' + str(datetime.date.today()), color="green", x=k*(datetime.date.today()-first_date).days, y=0, physics=False, shape='diamond', size=25, fixed={'x':True,'y':True})

    # if query_date is not None:
    #     g.add_node(str(query_date), label=' ', title='original date' + str(query_date), color="red", x=query_date_position, y=0, physics=True, shape='diamond', size=25, fixed={'x':True,'y':True})
    g.add_node("Old article", label=' ', title="Old article",color='yellow', x=0, y=0, physics=True, shape='diamond', size=35, fixed={'x':True,'y':True})



    # add node by node and 
    side = -1
    for c in clusters:
        g.add_node(c, label=filter_name(c), title=str(len(clusters[c]['links'])), size=30, mass=2, length=30*len(clusters[c]['links']), color='blue', y=side*30*len(clusters[c]['links']), fixed={'x':False, 'y':fix_level})

        if nodes[c][0] is not None:
            s=1
            for l2,_ in clusters[c]['links']:
                if nodes[l2][0] is not None and nodes[l2][0][:10] == nodes[c][0][:10]:
                    s += 1
            date = datetime.datetime.strptime(nodes[c][0][:10], '%Y-%m-%d').date()
            if first_date < date <last_date:
                g.add_edge(c, nodes[c][0][:10], color="red", width=s, title=str(s))
            elif date < first_date:
                g.add_edge(c, "Old article", color='yellow', physics=date_edge_physics, width=s, title=str(s))

        for link, source_type in clusters[c]['links']:
            if nodes[link][0] is not None:
                s=0
                date = datetime.datetime.strptime(nodes[link][0][:10], '%Y-%m-%d').date()
                for l2, _ in clusters[c]['links']:
                    if nodes[l2][0] is not None and nodes[l2][0][:10] == nodes[link][0][:10]:
                        s += 1
                if first_date < date < last_date:
                    g.add_edge(c, nodes[link][0][:10],title=str(s), label=source_type, physics=date_edge_physics, width=s)
                elif date < first_date:
                    g.add_edge(c, "Old article",title=str(s), label=source_type, color='yellow', physics=date_edge_physics, width=s)
                elif last_date <= date:
                    g.add_edge(c, nodes[link][0][:10],title=str(s), label=source_type, physics=date_edge_physics, width=s)
        side *= -1

    # add the nodes that have no children
    for n in nodes:
        if nodes[n][0] is not None:
            date = datetime.datetime.strptime(nodes[n][0][:10], '%Y-%m-%d').date()

            g.add_node(n, size=10, label=' ', title=n, y=side*30, x=max(0, k*(date-first_date).days), fixed={'x':True, 'y':True}, physics=False)
        side *=-1
    g.show_buttons()
    g.set_edge_smooth('smooth')
    g.show(f'Notebook graphs\\Timeline graphs\\2nd_version_{idx}.html')
```
